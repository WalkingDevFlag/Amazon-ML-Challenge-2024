```python
# Replace NaN with empty strings and convert ALL values to strings
df['Extracted Text'] = df['Extracted Text'].fillna('').astype(str)

# Verify fix
print(df['Extracted Text'].apply(type).unique())  # Should show [<class 'str'>]
```

```python
import torch
import pandas as pd
from transformers import BertTokenizer, BertModel
from tqdm import tqdm

def batch_generator(data, batch_size=16):
    """Yield successive batch-sized chunks from data."""
    for i in range(0, len(data), batch_size):
        yield data[i:i + batch_size]

def generate_bert_embeddings(df, text_column='Extracted Text', batch_size=16):
    # Clean and convert text data first
    df[text_column] = df[text_column].fillna('').astype(str)
    
    # Set up device and model
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased').to(device)
    model.eval()  # Set model to evaluation mode
    
    embeddings = []
    texts = df[text_column].tolist()
    
    with torch.no_grad():
        for batch in tqdm(batch_generator(texts, batch_size), 
                        total=len(texts)//batch_size + 1,
                        desc="Processing batches"):
            # Filter out empty/non-text entries
            valid_texts = [(idx, text) for idx, text in enumerate(batch) if text.strip()]
            
            if not valid_texts:
                embeddings.extend([None] * len(batch))
                continue
                
            # Unpack valid texts and indices
            indices, processed_texts = zip(*valid_texts)
            
            # Tokenize and process batch
            inputs = tokenizer(processed_texts, 
                             return_tensors='pt',
                             padding=True,
                             truncation=True,
                             max_length=512).to(device)
            
            outputs = model(**inputs)
            batch_embeddings = outputs.last_hidden_state.mean(dim=1).cpu().numpy()
            
            # Map embeddings back to original batch positions
            batch_results = [None] * len(batch)
            for i, idx in enumerate(indices):
                batch_results[idx] = batch_embeddings[i]
            
            embeddings.extend(batch_results)
    
    return pd.Series(embeddings, index=df.index)

# Usage
df = pd.read_csv(r'D:/Amazon ML Challenge 2024/outputs/output_paddleocr2.csv')
df['BERT_Embedding'] = generate_bert_embeddings(df, 
                                               text_column='Extracted Text',
                                               batch_size=32)  # Adjust based on GPU memory

# Save the extracted text and embeddings to a CSV file
csv_path = r'D:/Amazon ML Challenge 2024/outputs/train_extracted_text_with_embeddings.csv'
df.to_csv(csv_path, index=False)

print(f"Extraction and BERT encoding completed! The results have been saved to {csv_path}")
```
